### Grafik Akurasi

```python
method = ['Bagging', 'Bagging', 'Bagging', 'Boosting', 'Boosting', 'Boosting', 'Boosting', 'Boosting', 'Stacking']
model = ['Bagging', 'RF', 'ET', 'AdaBoost', 'GBM', 'XGBM', 'LGBM', 'CatBoost', 'Stacking']
accuracy = [81.17, 79.87, 79.22, 77.92, 81.82, 81.82, 79.22, 81.82, 81.17]

import pandas as pd

df_dataset1 = pd.DataFrame(data = {'method' : method, 'model' : model, 'accuracy' : accuracy})
```

```python
df_dataset1
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

ax = plt.figure(figsize=(10, 4), dpi=80)
ax = sns.barplot(x="model", y="accuracy", hue="method", data=df_dataset1, palette="tab10")
def change_width(ax, new_value) :
    for patch in ax.patches :
        current_width = patch.get_width()
        diff = current_width - new_value

        # we change the bar width
        patch.set_width(new_value)

        # we recenter the bar
        patch.set_x(patch.get_x() + diff * .5)
change_width(ax, .75)
plt.legend(loc = "lower right")
```
